```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import cv2
from skimage import io
from matplotlib.colors import NoNorm
```

```python
img = cv2.imread("/content/Desert.jpg")
cv2_imshow(img)
```

```python
r = np.arange(0,255)
c = 255/(np.log(1+255))
y = c*np.log(1+r)
plt.plot(r,y)

#Apply log transformation method

o=io.imread("/content/Desert.jpg")
c=255/(np.log(1+255))
log_image = c*np.log(o+1)

#float value will be converted to int

log_img = np.array(log_image, dtype= np.uint8)

plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
plt.imshow(o, cmap='gray', norm=NoNorm())
plt.subplot(1,3,2)
plt.imshow(log_img, cmap='gray', norm=NoNorm())
plt.show()
```

```python
r = np.arange(0,255)
c= 255/(np.log(1+255))
y = np.exp(r)**(1/c)-1
r = np.arange(0,255)
y = c*np.log(1+r)
c2 = 255/(np.log(1+255))
y2 = np.exp(r)**(1/c)-1
plt.plot(r,y);
plt.plot(r,y2,c2)

# Apply log inverse transformation
inv_log_image = np.exp(o**1/c)-1

#float value will be converted to int
inv_log_image = np.array(inv_log_image, dtype=np.uint8)
plt.figure(figsize =(15,4))
plt.subplot(1,3,1)
plt.imshow(o, cmap='gray', norm=NoNorm());
plt.subplot(1,3,2)
plt.imshow(inv_log_image, cmap='gray', norm=NoNorm())
plt.show()
```
