```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread("/content/Desert.jpg")
# img = cv2.resize(img,(250,170))
cv2_imshow(img)
```

```python
image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print("Image Shape",image.shape)
```

```python
# The image moves down, up, right, left

M = np.float32([[1,0,20],
               [0,1,100],
               [0,0,1]])

img1 = cv2.warpPerspective(image,M,(image.shape[1]*2, image.shape[0]*2))

M = np.float32([[1,0,20], [0,1,-100],[0,0,1]])
img2 = cv2.warpPerspective(image, M,(image.shape[1], image.shape[0]))


M = np.float32([[1,0,100],[0,1,20],[0,0,1]])
img3 = cv2.warpPerspective(image,M,(image.shape[1],image.shape[0]))

M = np.float32([[1,0,-100],[0,1,20],[0,0,1]])

img4 = cv2.warpPerspective(image, M, (image.shape[1], image.shape[0]))

#Display graphics 
titles=['orig', 'down', 'up', 'right', 'left']
images=[image, img1, img2, img3, img4]

for i in range(1,6):
  plt.subplot(2,3,1)
  plt.imshow(images[i-1])
  plt.title(titles[i-1])
  plt.xticks([]),plt.yticks([])
  plt.show()
```
