```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import cv2
from skimage import io
from matplotlib.colors import NoNorm
```

```python
# Image Saturation
img=cv2.imread('/content/download.jpg')
cv2_imshow(img)
```

```python
img_100 = np.ones(img.shape, dtype='uint8') * 100
cv2_imshow(img_100)
```

```python
img_add = cv2.add (img, img_100)
img_sub = cv2.subtract(img, img_100)
cv2_imshow(img_add)
cv2_imshow(img_sub)
```

```python
img_add1 = img_add + img_sub
cv2_imshow(img_add1)
```
